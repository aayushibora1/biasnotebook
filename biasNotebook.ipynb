```python
plt.rcParams["figure.figsize"] = [13.00, 5.50]
plt.rcParams["figure.autolayout"] = True
columns = ["Location", "dBLevel"]
plt.xticks(rotation = 300)
plt.ylabel("dB level in dBm")
plt.xlabel("Location")
plt.figtext(1, 1.08, "Phone Company: Xiaomi")
plt.figtext(1, 1.12, "Model Number: M1906G71")
plt.figtext(1, 1.16, "Processor:octa-core")
plt.figtext(1, 1.20, "RAM: 6GB")
plt.figtext(0.5, 1, "Bargraph between location and ASU")

plt.plot(spreadsheet.Location, spreadsheet.dBLevel)
```

```python
plt.rcParams["figure.figsize"] = [14.00, 7.50]
plt.rcParams["figure.autolayout"] = True
columns = ["Location", "UplinkFrequency","DownlinkFrequency"]
plt.xticks(rotation = 300)
plt.ylabel("Frequency")
plt.xlabel("Location")
plt.figtext(1, 1.08, "Phone Company: Xiaomi")
plt.figtext(1, 1.12, "Model Number: M1906G71")
plt.figtext(1, 1.16, "Processor:octa-core")
plt.figtext(1, 1.20, "RAM: 6GB")
plt.figtext(0.35, 1, "Linegraph between location and UplinkFrequency and DownlinkFrequency")

plt.plot(spreadsheet.Location, spreadsheet.UplinkFrequency,spreadsheet.DownlinkFrequency)
```
